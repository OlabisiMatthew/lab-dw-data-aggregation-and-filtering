2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
# Here, I am filtering the DataFrame for customers who responded "Yes" to the last marketing campaign
data = marketing_data[marketing_data['response'] == 'Yes']

# Grouping by 'policy_type' and 'gender' and calculating the average total_claim_amount
average_claim_amount = data.groupby(['policy_type', 'gender'])['total_claim_amount'].mean().reset_index()

print(average_claim_amount)
```

```python
#From the display above, the average claim amount for personal auto is higher the claim for other policy types, especially with the male gender. This shows that the incidence of accidents or other unfortunate circumstances is higher with with personal auto owned by males as compared to other males in other policy types. For the cars owned by females, the average claim amount for special auto is higher than other female in other policy category. . 
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# Grouping by state and also calculating the total number of customers in each state
data_by_state = marketing_data.groupby('state')['customer'].count().reset_index()

# Filtering data_by_state to only include states with more than 500 customers
data = data_by_state[data_by_state['customer'] > 500]

display(data)
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
# Here, I am grouping by education and gender and then calculate max, min, and median customer_lifetime_value
summary = marketing_data.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median']).reset_index()


display(summary)
```

```python
# The range of customer lifetime value is highest among the male gender whose educational level is high school or below. It can also be observed that the customer lifetime value is higher for females in every educational level apart from the high school or below category. 
```

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
# Create a pivot table to show the number of policies by state and month
pivot_table = pd.pivot_table(marketing_data, values='number_of_policies', index='state', columns='month', aggfunc='sum', fill_value=0)

display(pivot_table)
```

6.  Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
# total number of policies sold by state and month
total_policies = pd.pivot_table(marketing_data, values='number_of_policies', index=['state', 'month'], aggfunc='sum')

# top 3 states with the highest total number of policies sold
top3_states = total_policies.groupby('state').sum().nlargest(3, 'number_of_policies').index


df = marketing_data[marketing_data['state'].isin(top3_states)]


display(df)
```
